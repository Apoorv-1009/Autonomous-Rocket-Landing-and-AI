Q2. Given below is a list of dictionary, create a new list consisting of the values from the dictionary elements that are unique.
(1 mark)

```python
my_list=[
     {"first":"1"}, 
     {"second": "2"}, 
     {"third": "1"}, 
     {"four": "5"}, 
     {"five":"5"}, 
     {"six":"9"},
     {"seven":"7"}
    ]

# My solution

new_list = []
for dictionary in my_list:
  for value in dictionary.values(): # Iterates only once
    if value not in new_list:
      new_list.append(value)

# Desired list with unique values from the dictionaries
print(new_list)

```

Q16.
Debug the given below codes to generate a sin wave for 5 seconds duration. (7 Marks)

Note: Syntax error and Spelling mistakes at each line should be checked and corrected.

```python
import numpy as np
from matplotlib import pyplot as plt

SAMPLE_RATE = 44100  # Hertz
DURATION = 5  # Seconds

def generate_sine_wave(freq, sample_rate, duration):  # defining the function 
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)  # linespace to generate the sequence
    frequencies = x * freq
    # 2pi because np.sin takes radians
    y = np.sin((2 * np.pi) * frequencies)
    return x, y

# Generate a 2 hertz sine wave that lasts for 5 seconds
x, y = generate_sine_wave(2, SAMPLE_RATE, DURATION)
plt.plot(x, y)
plt.show();
```

```python
##Your Solution

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread("Image -ARL Assignment.jpg")

# Convert it to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Show the grayscale image
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap="gray")
plt.title('Original')

# Perform the canny edge detector to detect image edges
THRESHOLD1 = 50
THRESHOLD2 = 50

# Show the edges of the image
edges = cv2.Canny(gray, THRESHOLD1, THRESHOLD2)
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap="gray")
plt.title('Edge detected image')
plt.show()

# My comments
# Smallest value between threshold1 and threshold2 is used for edge linking and 
# the largest value is used to find initial segments of strong edges.
# THRESHOLD1 value sets the edge sensitivity along the x axis, while THRESHOLD2 
# sets the sensitivity along the y axis
```

6. Call your method on the person you created. Make sure your method executed properly; if the method does not print anything out directly, print something before and after calling the method to make sure it did what it was supposed to. (3)

```python
##Your Solution

class Person:
  def __init__(self, name, age, pob, dob, language):
    self.name = name
    self.age = age
    self.pob = pob
    self.dob = dob
    self.language = language

  def introduce_yourself(self):
    print("Hello, my name is", self.name)
    print("I am", self.age, "years old")
    print("I was born in", self.pob, "on", self.dob)
    print("My mother toungue is", self.language)

  def age_person(self):
    self.age += 1
    print("I am now", self.age, "years old")

p = Person("Apoorv", 21, "Delhi", "10/09/2001", "Hindi")
p.introduce_yourself()
p.age_person()
```
