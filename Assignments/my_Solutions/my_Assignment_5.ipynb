#Question 1 (10 Marks)

**"Differentiate Value Based methods and Policy Based methods in brief"**

Value-based methods and policy-based methods are two broad categories of reinforcement learning algorithms.

There are 2 broad categories of reinforcement learning:

Value-based methods: In this method we aim to find the optimal value function or the optimal state-action values. The value function estimates how good a particular state or state-action pair is in terms of the expected cumulative future reward. Examples of value-based methods include Q-learning. These methods do not explicitly search for the optimal policy, but they can derive it from the value function.

Policy-based methods: Here we aim to directly find the optimal policy. Instead of estimating the value function, policy-based methods directly optimize the policy, which is a mapping from states to actions. An example of policy-based method is the actor-critic method. Policy-based methods typically work well in continuous action spaces and can directly optimize stochastic policies.

#Question 2 (15 marks)

**"Briefly explain intuition of Actor Critic Method with drawbacks"**

In the actor-critic method, there are two main components - the "actor" and the "critic". The actor makes decisions based on the current state of the environment, while the critic evaluates those decisions to provide feedback.

The actor is like a student who takes an exam. It receives information about the environment, such as the current state, and then makes a decision about what action to take next. This decision is based on what it has learned through past experiences.

The critic is like a teacher who grades the student's exam. It evaluates the decision made by the actor and provides feedback on how good or bad it was. This feedback is then used to improve the actor's decision-making process in the future.

One of the benefits of the actor-critic method is that it can learn from both positive and negative feedback. This allows it to improve its decision-making abilities over time.

However, there are also some drawbacks to this approach. One is that it can be difficult to tune the parameters of the algorithm. Additionally, it can be prone to overfitting, which means that it may perform well on the training data but not on new, unseen data. Finally, it may be computationally expensive, especially when working with large datasets.

```
import gymnasium as gym
import numpy as np

from stable_baselines3 import DQN   # Deep Q-Network (DQN)
from stable_baselines3.common.evaluation import evaluate_policy
```

## Step 2: Create the Gym env and instantiate the agent

```
model = DQN(
    "MlpPolicy",   # MlpPolicy (Multi-Layer Perceptron) is chosen because the input of lunar lander is a vector
    "LunarLander-v2",
    verbose=1,   # To print updates and progress information
    exploration_final_eps=0.1,   #  Sets the final value of the exploration epsilon parameter
    target_update_interval=250,   # Sets the frequency at which the target network used in the DQN algorithm is updated
)
```

## Step 3: Evaluate the un-trained agent

```
# Separate env for evaluation
env = gym.make("LunarLander-v2")

# Agent before training
mean_reward, std_reward = evaluate_policy(
    model,
    env,
    n_eval_episodes=10,
    deterministic=True,
)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

## Step 4: Train the agent and save it



```
# Train the agent
model.learn(total_timesteps=int(1e5))
# Save the agent
model.save("dqn_lunar")
```

## Step 5: Load the trained agent and obtain Mean Reward

```
model = DQN.load("dqn_lunar")

# Evaluate the trained agent
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10, deterministic=True)

print("mean_reward=", mean_reward, " +/- ", std_reward)
```
