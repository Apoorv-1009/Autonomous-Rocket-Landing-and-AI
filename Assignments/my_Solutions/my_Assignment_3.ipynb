# Ans
The actual y_min for the function can be found by taking the derivative of the given function: $ y = x^{2} -4x + 1 $ <br>
Taking the first derivative with respect to x, we get: $ \frac {dy}{dx} = 2x - 4 $ <br>
Now we set the derivative equal to 0 to find the point where the slope of the function is 0, thus giving us the x-coordinate where y is minimum. <br>
Thus we have: $ 2x - 4 = 0 $ <br>
and $ x = 4/2 = 2 $ <br> 
Thus at x = 2, y will have its minimum value which can be calculated using its function as: y_min = $ 2^2 -4*2 + 1 = -3 $

````python
#Ans

# Import libraries 
import numpy as np
import matplotlib.pyplot as plt
import math 

# Define the function y = x^2 - 4x + 1
def function(x):
  return x**2 - 4*x + 1

# Define the derivative of the function 
def derivative(x):
  return 2*x - 4

# Define the gradient descent algorithm
def gradient_descent(x_start, learning_rate, iterations, tolerance):
  x = x_start
  x_history = [x]
  converge = False

  for i in range(iterations):
    gradient = derivative(x)
    delta_x = -learning_rate * gradient
    x += delta_x
    x_history.append(x)
    if abs(delta_x) < tolerance : # If growth is below threshold
      converge = True
      break
  return x, x_history, converge, i

# Define hyperparameters
starting_value_of_x = 10
iterations = 100
learning_rates = [0.01, 0.05, 0.1, 0.4, 0.5, 0.9, 0.99, 0.999, 1]
tolerance = 1e-3    

# Define rows for subplots and visualisation
rows = 3
cols = math.ceil(len(learning_rates) / rows)
fig, axs = plt.subplots(rows, cols, figsize=(17, 17))
i, j = 0, 0

# Define x and y values of function for plotting
x = np.linspace(-7, 12, 1000)
y = function(x) 

for learning_rate in learning_rates:
  x_min, x_history, converge, iters = gradient_descent(starting_value_of_x, learning_rate, iterations, tolerance)

  # Plot the function for easier visualization
  axs[i, j].plot(x, y)

  # Plot the gradient descent iterations
  y_history = function(np.array(x_history))
  axs[i, j].plot(x_history, y_history, linestyle='dashed', color='red', marker='o', label='learning rate = {}'.format(learning_rate))
  axs[i, j].set_title('Convergence: {} \n Iterations: {}, xmin: {} \n ymin: {}'.format(converge, iters+1, x_min, function(x_min)))

  axs[i, j].set(xlabel='x-label', ylabel='y-label')
  axs[i, j].legend()
  axs[i, j].grid()

  # Update to next subplot
  j += 1  # Next column in row
  if j == cols: # When reached last column in row
    j = 0 # Reinitialise column
    i += 1  # Go to next row

plt.subplots_adjust(wspace=0.5, hspace=0.5, top=0.85)
plt.show()
````

From the given graphs, we can infer:
1. With learning rate = **0.1**, the model converges very slowly to the minimum value, thus demanding more iterations.
2. Increasing the learning rate from **0.1**, it is converging close to the minimum at a much faster rate, thus requiring lesser iterations to reach the global minima.
3. With the learning rate = **0.5**, the model requires the least amount of iterations and is able to reach the most accurate value of the global minima.
4. Further increasing the learning rate from **0.5**, the model starts to take more iterations to reach the global minima before being unable to reach anywhere near the minima entirely.
5. At higher values of the learning rate, the model does not converge to the minima but rather increases in value before spiraling out of control.

Thus for the given question and our set of hyperparameters the most optimal learning rate is **0.5**, taking 2 iterations to reach the minima with the most accurate value.

#Question 4: (5 Marks)

**List Out various Optimizers used in Deep Neural Network and briefly explain about Adam Optimizer**

# Ans
Here are a few optimizers used in neural networks:

1. **Stochastic Gradient Descent (SGD)**: This optimizer updates the weights of the neural network in the direction of the negative gradient of the loss function.

2. **Adaptive Moment Estimation (Adam)**: Uses a combination of momentum and adaptive learning rates. Here, it keeps track of the first and second moments of the gradients to adjust the learning rate dynamically during training.

3. **Root Mean Square Propagation (RMSprop)**: Here we use the moving average of the squared gradients to adjust the learning rate. It scales the learning rate based on the recent gradient magnitudes.

4. **Adagrad**: This optimizer adapts the learning rate of each parameter based on the historical gradients for that parameter. 

5. **Adadelta**: It is an optimizer that extends Adagrad to reduce its aggressive, monotonically decreasing learning rate by using a moving window of the gradient updates to adjust the learning rate.

6. **Nesterov Accelerated Gradient (NAG)**: It is an optimizer that uses the momentum technique to accelerate SGD by calculating the gradient ahead of time using the current velocity and uses it to update the weights.

Speaking specifically about the **Adam Optimizer**, this optimizer combines the concepts of momentum and adaptive learning rates. Adam optimizer maintains a moving average of the first and second moments of the gradients to adaptively adjust the learning rate during training. It uses the gradient information from the current batch of data to update the parameters, and the moving averages to adjust the learning rate for each parameter separately. The adaptive learning rates help the optimizer to converge faster, and the momentum helps the optimizer overcome local minima and saddle points in the overall loss landscape. Adam optimizer is widely used in deep learning and is known to be efficient in a wide range of neural network architectures and applications.

Build a small binary Classification architecture using Convolutional Neural Networks.

Things to add and take care:

2. The total trainable parameters should be more than 20,000,000.
3. Input Image size: (224, 224, 3)

````python
#Ans
from tensorflow.keras.models import Sequential, Model 
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Input
````

# Notes:
1. The rectified linear activation function or ReLU for short is a piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero. It is defined as f(x) = max(0, x), which means that any negative input is set to zero, while positive inputs are passed through unchanged.
2. Pooling layers are used to downsample the output of convolutional layers by summarizing their activations. This helps in reducing the spatial dimensions of the input, thereby reducing the number of parameters and computation required in the network. The most commonly used pooling operation is max pooling, where the maximum value in each non-overlapping window of the feature map is selected and passed on to the next layer. <br>
The basic idea of **maxpooling** is to divide the input image or feature map into non-overlapping rectangular regions (called pooling regions or filters) and for each region, output the maximum value. The size of the pooling regions (pool size) is a hyperparameter that can be tuned during model training. Typically, a pool size of (2, 2) or (3, 3) is used.

````python
model = Sequential()
filters = 1024
i = 0

# First convolution layer 
# filters number of filters , 3x3 filter size
model.add(Conv2D(filters/(2**i), (3, 3), activation="relu", input_shape=(224,224, 3)))
model.add(MaxPooling2D((2, 2)))
i+= 1

# Second convoluion layer
# filters/2 number of filters, 3x3 filter size
model.add(Conv2D(filters/(2**i), (3, 3), activation="relu"))
model.add(MaxPooling2D((2, 2)))
i+= 1

# Third convolution layer
# filters/4 number of filters, 3x3 filter size
model.add(Conv2D(filters/(2**i), (3, 3), activation="relu"))
model.add(MaxPooling2D((2, 2)))
i+= 1

# Flatten layer to convert the 3D feature maps to 1D feature vectors
model.add(Flatten())

# Fully connected layer with filters/8 number of filters and ReLU activation function
model.add(Dense(filters/(2**i), activation='relu'))

# Output layer with a single sigmoid activation unit for binary classification
model.add(Dense(1, activation='sigmoid'))

# Print summary
model.summary()
````
