#Question 2: (20 Marks)

**Find the value of x for which $y = x^2 - 4x + 1$ is minimum using the gradient decent algorithm.**              
**Visualise the process of convergence using different learning rates during a total of 100 iterations.**

*Use the following hyperparameters for gradient descent:-*

```
starting_value_of_x = 10

learning_rates = [0.01, 0.05, 0.1, 0.4, 0.5, 0.9, 0.99, 0.999, 1]
tolerance = 1e-3    # Here, tolerance is the condition to be used to note if the model is converging to its minimum point or not
```

**Requirments of this question for each learning rate:-**         
- *Manually calculate the value of x for which y is minimum and show it in the output screen*     
- *How many iterations did the model take to converge?*       
- *If the model didn't converge then what is the value of x after 100 iterations?*        
- *Generate the subplots of the function with all the points in the process of convergence for each learning rate*  
- *Keep the track of such points in the plots*
- *Write your final conclusions about the effect of learning rates on convergence of this problem*

**Below is one example to help you understand the requirments:-**        
<img src = 'https://drive.google.com/uc?id=1Aga0qLfxxqjHAlatDaVfJoqrn61pb65X'>

````python
#Answer
import numpy as np
plt.style.use('default')

''' PREPARING SOME FUNCTIONS '''

# Main equation
def f(x):
  return x**2-4*x+1

# Gradient equation
def gradient_func(x):
  return 2*x - 4

#  Gradient Descent
def gradient_descent(start, gradient_eqn, learn_rate, max_iter, tol):

  steps = [start]                                                               # history tracking
  x = start

  for _ in range(max_iter):
    diff = learn_rate*gradient_eqn(x)
    if np.abs(diff)<tol:
      break    
    x -= diff                                                                   # x_new = x_prev - diff
    steps.append(x)                                                             # history tracing

  return steps, x

# Visualising
def graph_GD(history, max_iter, result):

  # Create x and y data to plot the eqn
  x = np.arange(-7, 11, 0.2) 
  y = f(x)
  
  # Create respective y position for x values in history
  y_pos = [f(x) for x in history]

  # Ploting begins
  plt.plot(x, y, 'royalblue')
  plt.plot(history, y_pos, 'ro--', markeredgecolor = 'm', markerfacecolor = 'y', markersize = 5)
  if len(history)-1 == max_iter:
    plt.title(f"Couldn't converge at learning rate of {lr}\nAfter {len(history)-1} interations, x = {round(result,3)} (3sf)", y = -0.15, color = 'green')
  else:
    plt.title(f"Took {len(history)-1} iterations to converge\nat a learning rate of {lr}", y = -0.15, color = 'green')
  plt.axis('off')
````

````python
''' DRIVER CODE '''

# Header 
print(f"\t\t{'*'*81}")
print("\t\t*\tVisualising the effect of learning rates during 100 iterations\t\t*")
print("\t\t*\t\t\tx = 2 is the point of convergence\t\t\t*")
print(f"\t\t{'*'*81}")

# Hyperparameters
start = 10
max_iter = 100
alpha = [0.01, 0.05, 0.1, 0.4, 0.5, 0.9, 0.99, 0.999, 1]
tol = 1e-3

# Subplots to check the effect of learning rate during 100 iterations
plt.figure(figsize = (15,15))
plt.style.use('dark_background')  # Run code cell twice to see this effect on colab / could have also set backgroud to black
for idx, lr in enumerate(alpha):
  history, result = gradient_descent(start, gradient_func, lr, max_iter, tol)
  plt.subplot(3, 3, idx+1)
  graph_GD(history, max_iter, result)
plt.subplots_adjust(hspace = 0.3)
plt.show()
````

**Conclusions:-**        
**1)** *Much lower learning rates slowed down the process of convergence.*           
**2)** *Learning rates closer to 1 were again taking time to converge.*        
**3)** *Learning rate of 1 caused the divergence where the values were simply toggling between -10 and 10.*    
**4)** *Learning rate of 0.5 gave the quickest convergence.*            

*Learning rate must not be too high or too low! We saw that how changing the learning rates were impacting the learning process. Hence, it is recommended to try some of the learning rates with some callbacks (will learn about it in later sessions) to efficiently complete the learning process.*

Build a small binary Classification architecture using Convolutional Neural Networks.

Things to add and take care: 

1. 3 Convolutional layers with MaxPooling
2. The total trainable parameters should be more than 20,000,000.
3. Input Image size: (224, 224, 3)

````python
#Ans
````

````python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense, BatchNormalization, GlobalMaxPooling2D
from keras import backend as K

def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true))) 

model = Sequential()
model.add(Conv2D(2058, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(Conv2D(1024, (3, 3), activation='relu',))
model.add(Conv2D(256, (3, 3),activation='relu')) 
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dense(128, activation='relu')) 
model.add(Dropout(0.25)) 
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adamax', metrics=[root_mean_squared_error])
model.summary()
````
