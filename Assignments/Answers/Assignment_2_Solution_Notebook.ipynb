#Question 6 (15 Marks)

Use SymPy to define the fuel loss function as follows,and covert those into numerical python function then plot different ways to expell fuel with repect to 1000 finite timescale (0.0001 to 1) and n = (0.5,1,1.5,2,3,5,7,10).Finally justify which Fuel Loss function is optimal to obtain stable and controlled flight.

image.png



```python
#Ans
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import sympy as smp

plt.style.use(['seaborn-bright', 'seaborn-dark'])

#Defining all required constants
T0 = 40 # 40 second launch 
g = 9.81
vg = 500 # fuel emitted at 500m/s
M0 = 2e4 #20000kg rocket starting
b = 0.1 # coefficient of air friction
m0 = M0/(b*vg*T0)

#Setting up variables in sympy
t, n = smp.symbols('t n')
z = 1-(0.7)*t**n
dzdt = smp.diff(z, t).simplify()

#Lambdifying them for use in python
z = smp.lambdify([t, n], z)
dzdt = smp.lambdify([t, n], dzdt)

t = np.linspace(1e-4, 1, 1000)

n_var_list = [0.5, 1, 1.5, 2, 3, 5, 7, 10] #A list of every n value given
z_func_list = [] #A list of z function values based on n 

#Looping through n values and appending z function values in the list
for n_var in n_var_list:
    z_func_list.append(z(t, n_var))

plt.figure(figsize=(15,7))

#Plotting z values onto the graph
for idx, z_func in enumerate(z_func_list):
    plt.plot(t, z_func, label = f"M{idx + 1}")

plt.ylabel('$m(t)/m_0$')
plt.xlabel('$t/T_0$')
plt.legend()
plt.show()
```

In the above graph, we have eight different fuel loss function plots. Each plot uses a different "n" value as per the given criteria. M1 to M8 are final values of our "z" function equation, using the respective n values in order.

Fuel Loss Function M1(t, 0.5) is showing that fuel gradually depletes in a underside curve until the fuel tank is empty. This is done in a 0.001 to 1, 1000 finite timescale.
Fuel Loss Function M2(t, 1) is a linear slope, decreasing fuel in almost a constant rate of discharge.
Fuel Loss function M3 to M5 is gradually decreasing the fuel but in an upper curve, which get's steeper after each increase of n value.
Fuel Loss Function M6 to M8 are showing that fuel depletion is very small for a while, before it suddenly starts depleting fuel at a high rate until it's empty.
