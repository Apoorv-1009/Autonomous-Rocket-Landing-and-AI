**Linear Regression**

The objective of a linear regression model is to find a relationship between one or more features(independent variables) and a continuous target variable(dependent variable). When there is only feature it is called Uni-variate Linear Regression and if there are multiple features, it is called Multiple Linear Regression.

**Simple Linear Regression** using Least Square method:

**Least squares** is a statistical method used to determine the best fit line or the regression line by minimizing the sum of squares created by a mathematical function. The “square” here refers to squaring the distance between a data point and the regression line. The line with the minimum value of the sum of square is the best-fit regression line

Regression Line, y = mx+c where,

y = Dependent Variable

x= Independent Variable ; c = y-Intercept

Screenshot 2022-01-27 141016.png

```python
#finding values for m and c
num = 0
den = 0
n = len(X)
for i in range(n):
  num += (X[i] - np.mean(X)) * (Y[i] - np.mean(Y))
  den += (X[i] - np.mean(X)) ** 2
m = num/den
c = np.mean(Y) - (m * np.mean(X))    # y = mx+c , So c = y-mx
print(m,c)
```

```python
#plotting the regression line
plt.scatter(X, Y, color = "#ef5423", label = 'Scatter Plot')  #scatter plot
y = m*X + c  
#plt.scatter(X,y, label = 'Regression Line') #regression line
plt.plot(X,y, label = 'Regression Line') #regression line
plt.legend()
```

#Calculating R2 score

image.png

```python
#calculating R-square
#ss_t is the total sum of squares and ss_r is the total sum of squares of residuals(relate them to the formula).
ss_t = 0
ss_r = 0
for i in range(n):
  y_pred = c + m * X[i]
  ss_t += (Y[i] - np.mean(Y)) ** 2            
  ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)
print(r2)
```

R2 with 0.75, 0.50, and 0.25 are described as substantial, moderate and weak respectively.

**Linear Regression using Gradient Descent**

```python
#gradient descent
m = 0.1
c = 0.1

L = 0.1  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # 100
# Number of elements in X
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y # Y_pred is of 100 instance
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c   
print (m, c)
```

```python
#plotting the regression line
plt.scatter(X, Y, color = "#ef5423", label = 'Scatter Plot')  #scatter plot
y = m*X + c
plt.plot(X,y, label = 'Regression Line') #regression line
plt.legend()
```
